```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from src.config.settings import RESULTS_PATH
```

plotting functions

```python
def barplot(df_array, metric, labels):
    losses = [df_array[i][metric][0] for i in range(len(labels))]
    plt.scatter(y=losses, x=np.arange(len(labels)),marker='x', label=metric, alpha=1)
    plt.xticks(np.arange(len(labels)), labels, rotation=45);

```

import tests results

```python
runs_english = {
    'bag of words':'ENGLISH_BOW_1',
    'nearest mean (length)': 'ENGLISH_NM_1',
    'Word Embeddings': 'ENGLISH_W2V_14',
    'Document Embeddings':'ENGLISH_D2V_1'
}

runs_japanese = {
    'bag of words':'JAPANESE_BOW_3',
    'nearest mean (length)': 'JAPANESE_NM_1',
    'Word Embeddings': 'JAPANESE_W2V_14',
    'Document Embeddings':'JAPANESE_D2V_6'
}

english_dfs = []
japanese_dfs = []
losses = {}
for run in runs_english.keys():
    english_dfs.append(pd.read_csv(os.path.join(RESULTS_PATH, 'test', runs_english[run], 'result_metrics.csv')))
    japanese_dfs.append(pd.read_csv(os.path.join(RESULTS_PATH, 'test', runs_japanese[run], 'result_metrics.csv')))
```

### English Tests Results

```python
barplot(english_dfs, 'loss', runs_english.keys())
plt.title('loss')
plt.show();
plt.title('metrics')

barplot(english_dfs, 'avg_accuracies', runs_english.keys())
barplot(english_dfs, 'avg_f1s', runs_english.keys())
barplot(english_dfs, 'avg_precisions', runs_english.keys())
plt.title('metrics')
plt.legend();
```

### Japanese Test Results

```python
barplot(japanese_dfs, 'loss', runs_japanese.keys())
plt.title('loss')
plt.show()
barplot(japanese_dfs, 'avg_accuracies', runs_japanese.keys())
barplot(japanese_dfs, 'avg_f1s', runs_japanese.keys())
barplot(japanese_dfs, 'avg_precisions', runs_japanese.keys())
plt.title('metrics')
plt.legend();
```

All models seem to perform similarly exept for the NM model which perform significantely worse, The word embedding shows a slight advantage wrt the other models.
